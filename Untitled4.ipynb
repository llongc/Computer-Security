```python
import numpy as np
message = ''
with open('ciphertext-sample.txt','r') as f:
    for line in f:
        message += line.lower()
```

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
count = np.zeros(26)
msg = []
diction = []
```

```python
with open('dictionary-sample.txt','r') as dic:
        for line in dic:
            diction.append(line[:-1])
```

```python

for key in range(len(letters)):

    rst = ''
    for symbol in message:
    
        if symbol in letters:
        
            num = letters.find(symbol)
            shift = num + key
            if shift > 25:
                shift = shift - 26
            rst = rst + letters[shift]
        else:
            rst = rst + symbol

    msg.append(rst)
    for word in rst.split():
    
        if word in diction:
            
            count[key] = count[key] + 1

```
