```python
def freqAnalysis(FileName):
    letters = 'abcdefghijklmnopqrstuvwxyz '
    count = []
    msg = ''
    w = 0
    for i in range(27):
        count.append(0)
    file = open(FileName, "r")

    for line in file:
        #line = line.lower()
        #print line
        for x in line:
            x = x.lower()
            msg += x
            i = 0
            if x in letters:
                count[letters.find(x)] += 1
    file.close()
    return count,msg
```
